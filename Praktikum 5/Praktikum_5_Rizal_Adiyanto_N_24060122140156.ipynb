## Import

```python
import numpy as np
import pandas as pd
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from matplotlib import pyplot as plt
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
X1, y1 = make_blobs(n_samples=50, centers=[[4,4], [-2,-1], [1,1], [10,4]], cluster_std=0.9)
plt.scatter(X1[:, 0], X1[:, 1], marker='o')
```

## Dataset Random

```python
single_linkage = AgglomerativeClustering(n_clusters=4, linkage='single')
y_single = single_linkage.fit_predict(X1)

average_linkage = AgglomerativeClustering(n_clusters=4, linkage='average')
y_average = average_linkage.fit_predict(X1)
```

Kode ini menggunakan algoritma Agglomerative Clustering untuk mengelompokkan data X1 menjadi 4 klaster dengan dua pendekatan yang berbeda. Pendekatan pertama menggunakan single linkage, yaitu menggabungkan klaster berdasarkan jarak terdekat antara dua titik. Hasil klasterisasi dari pendekatan ini disimpan dalam variabel y_single. Pendekatan kedua menggunakan average linkage, yaitu menggabungkan klaster berdasarkan rata-rata jarak antara semua titik dari dua klaster. Hasilnya disimpan dalam variabel y_average.

```python
plt.scatter(X1[:, 0], X1[:, 1], c=y_single, cmap='rainbow', marker='o', label='Single Linkage')
plt.title("Agglomerative Clustering - Single Linkage")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(["Cluster Assignment"])
plt.show()
```

Menampilkan hasil klustering menggunakan Single Linkage

```python
plt.scatter(X1[:, 0], X1[:, 1], c=y_average, cmap='rainbow', marker='o', label='Average Linkage')
plt.title("Agglomerative Clustering - Average Linkage")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(["Cluster Assignment"])
plt.show()
```

Menampilkan hasil klustering menggunakan Average Linkage

## Dataset Cars

```python
filename = 'cars_clus.csv'

pdf = pd.read_csv(filename)
print("Shape of dataset", pdf.shape)

pdf.head(20)
```

Import dataset cars dan menampilkan 20 row pertama dataset

```python
print("Shape of dataset before cleaning", pdf.size)
pdf[['sales', 'resale', 'type', 'price', 'engine_s',
     'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap']] = pdf[['sales', 'resale', 'type', 'price', 'engine_s',
     'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap']].apply(pd.to_numeric, errors='coerce')
pdf = pdf.dropna()
pdf = pdf.reset_index(drop=True)
print("Shape of dataset after cleaning", pdf.size)
pdf.head(5)
```

Mengkonversi beberapa kolom seperti 'sales', 'resale', 'type', 'price', 'engine_s', 'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap' menjadi tipe numerik dengan pd.to_numeric, menggunakan parameter errors='coerce' untuk menggantikan nilai yang tidak bisa dikonversi dengan NaN.

Baris yang mengandung NaN dihapus dengan fungsi dropna(), dan indeks dataset diatur ulang menggunakan reset_index(drop=True).

```python
featureset = pdf[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
print(featureset.head(5))
```

```python
from sklearn.preprocessing import MinMaxScaler
x = featureset.values
min_max_scaler = MinMaxScaler()

featureset = featureset.apply(pd.to_numeric, errors='coerce').dropna()
x = featureset.values

feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx [0:5]
```

Kode ini digunakan untuk melakukan normalisasi pada dataset featureset menggunakan metode Min-Max Scaling. Langkah pertama adalah memastikan data bersih dengan mengubah seluruh nilai dalam dataset menjadi tipe numerik menggunakan apply(pd.to_numeric, errors='coerce'), di mana nilai yang tidak dapat dikonversi akan diganti dengan NaN, dan baris yang mengandung NaN dihapus menggunakan dropna(). Setelah itu, nilai-nilai dalam dataset diambil dalam bentuk array menggunakan featureset.values dan disimpan dalam variabel x. Objek MinMaxScaler kemudian dibuat untuk melakukan normalisasi, yang mengubah semua nilai fitur ke dalam rentang 0 hingga 1 menggunakan fungsi fit_transform(). Hasil normalisasi disimpan dalam variabel feature_mtx, dan lima baris pertama dari matriks tersebut ditampilkan untuk memverifikasi hasilnya.

```python
single = AgglomerativeClustering(n_clusters=6, linkage='single')
single.fit(feature_mtx)
print("Labels using Single Linkage (Scikit-learn):")
print(single.labels_)

average = AgglomerativeClustering(n_clusters=6, linkage='average')
average.fit(feature_mtx)
print("\nLabels using Average Linkage (Scikit-learn):")
print(average.labels_)
```

Melakukan klustering data yang sudah dinormalisasi dan disimpan pada feature_mtx dengan pendekatan Single Linkage dan Average Linkage. Model ini diatur untuk membagi data menjadi 6 klaster (n_clusters=6) dan dilatih menggunakan fit(feature_mtx).

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z_single = linkage(feature_mtx, method='single')
Z_average = linkage(feature_mtx, method='average')

# Visualisasi Dendrogram untuk Single Linkage
plt.figure(figsize=(10, 7))
dendrogram(Z_single)
plt.title("Dendrogram (Single Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# Visualisasi Dendrogram untuk Average Linkage
plt.figure(figsize=(10, 7))
dendrogram(Z_average)
plt.title("Dendrogram (Average Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

Pembuatan dendogram untuk hasil klustering, masing masing hasil disimpan dalam Z_single untuk single linkage dan Z_average untuk average linkage. Kemudian dendogram dibuat dengan fungsi dendrogram(Z_single) dan dendrogram(Z_average).

## Dataset Iris

```python
import pandas as panda
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
dataset = panda.read_csv(url, names=names)
```

Import dataset

```python
print(dataset.shape)
print(dataset.head(100))
```

```python
features = dataset[['sepal-length', 'sepal-width', 'petal-length', 'petal-width']]
target = dataset['Class']
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Single Linkage Clustering using Scikit-learn
single_linkage = AgglomerativeClustering(n_clusters=3, linkage='single')
single_labels = single_linkage.fit_predict(features_scaled)

# Average Linkage Clustering using Scikit-learn
average_linkage = AgglomerativeClustering(n_clusters=3, linkage='average')
average_labels = average_linkage.fit_predict(features_scaled)

# Complete Linkage Clustering using Scikit-learn
complete_linkage = AgglomerativeClustering(n_clusters=3, linkage='complete')
complete_labels = complete_linkage.fit_predict(features_scaled)

# Print the labels for each method
print("\nSingle Linkage Labels:")
print(single_labels)

print("\nAverage Linkage Labels:")
print(average_labels)

print("\nComplete Linkage Labels:")
print(complete_labels)
```

menggunakan algoritma Agglomerative Clustering dari Scikit-learn untuk membagi data yang telah dinormalisasi dalam features_scaled menjadi 3 klaster, dengan tiga pendekatan linkage yang berbeda.

```python
Z_single = linkage(features_scaled, method='single')
plt.figure(figsize=(10, 7))
dendrogram(Z_single)
plt.title("Dendrogram (Single Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

Z_average = linkage(features_scaled, method='average')
plt.figure(figsize=(10, 7))
dendrogram(Z_average)
plt.title("Dendrogram (Average Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

Z_complete = linkage(features_scaled, method='complete')
plt.figure(figsize=(10, 7))
dendrogram(Z_complete)
plt.title("Dendrogram (Complete Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

Pembuatan dendogram untuk masing masing hasil klustering
